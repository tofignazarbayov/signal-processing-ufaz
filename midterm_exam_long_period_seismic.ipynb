# Question 3

```python
#load data
fname = 'ECH.G.ascii.dat'
lhz_data = np.loadtxt(fname, usecols=0)
```

```python
dt = 1
time = np.arange(0, len(lhz_data)) * dt
lhz_data = lhz_data - np.mean(lhz_data)

lhz_DATA = fft.fft(lhz_data)
N = len(time)
f = fft.fftfreq(N, dt)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(f[0:int(N/2)], np.abs(lhz_DATA[0:int(N/2)]) / 1e6)
plt.plot()
plt.xlabel('frequency (s-1)')
plt.title('amplitude spectrum')
plt.xlim(5 / 1000000, 30 / 1000000)
plt.ylim(0, 1500)
plt.grid()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(f[0:int(N/2)], np.angle(lhz_DATA[0:int(N/2)]))
plt.xlabel('frequency (s-1)')
plt.title('phase spectrum')
plt.xlim(5 / 1000000, 30 / 1000000)
plt.grid()
```

# Question 5

```python
#load data
fname = 'ECH.G.ascii.dat'
lhz_data = np.loadtxt(fname, usecols=0)
```

```python
dt = 1
time = np.arange(0, len(lhz_data)) * dt / (60 * 60 * 24)
time = [i for i in time if i < 14]
lhz_data = lhz_data - np.mean(lhz_data)
lhz_data = lhz_data

lhz_DATA = fft.fft(lhz_data)
N = len(time)
f = fft.fftfreq(N, dt)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(f[0:int(N/2)], np.abs(lhz_DATA[0:int(N/2)]) / (0.5 * 1e6))
plt.plot()
plt.xlabel('frequency (s-1)')
plt.title('amplitude spectrum (14 days)')
plt.xlim(5 / 1000000, 30 / 1000000)
plt.ylim(0, 2300)
plt.grid()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(f[0:int(N/2)], np.angle(lhz_DATA[0:int(N/2)]))
plt.xlabel('frequency (s-1)')
plt.title('phase spectrum (14 days)')
plt.xlim(5 / 1000000, 30 / 1000000)
plt.grid()
```

```python
import numpy as np
import math
from scipy import signal
from numpy import fft
import matplotlib.pyplot as plt
```

Your data today is the file `sunspots.dat` which contains a record of timeseries of number of sunspots:
* Column 1 : year of observation
* Column 2 : Number of  black spots observed on the sun (sunspots)

```python
#load data
A = np.loadtxt('sunspots.dat')
# variables
year = A[:,0]
spots = A[:,1]
```

## Question 1. Plot the data in the time domain

Your data shoud be represented with labels, legend and title.

```python
plt.plot(year, spots, label = "spots")
plt.xlabel("year")
plt.ylabel("yearly number")
plt.title("timeseries of sunspots")
plt.legend()
```

## Question 2. Plot the data in the frequency domain

Your data shoud be represented with labels, legend and title.

```python
dt = 1
N = len(spots)
spots = spots - np.mean(spots)
SPOTS = fft.fft(spots)
f = fft.fftfreq(N,dt)

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(f[0:int(N/2)],np.abs(SPOTS[0:int(N/2)]), label = "amp")
plt.xlabel("frequency (y-1)")
plt.title("sunspots : amplitude spectrum")
plt.xlim((0,1/(2*dt)))
plt.grid()
plt.legend()

plt.subplot(122)
plt.plot(f[0:int(N/2)],np.angle(SPOTS[0:int(N/2)]), label = "phase")
plt.xlabel("frequency (y-1)")
plt.title("sunspots : phase spectrum")
plt.xlim((0,1/(2*dt)))
plt.grid()
plt.legend()
```
